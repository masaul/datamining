```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.compose import make_column_transformer
```

### Insert Dataset

```python
  data = pd.read_csv("https://raw.githubusercontent.com/masaul/data-csv/main/credit_score.csv")
  data
```

### Convert ke dataframe

```python
df_without_column_for_convert = pd.DataFrame(data, columns = ['kode_kontrak','pendapatan_setahun_juta','durasi_pinjaman_bulan','jumlah_tanggungan','risk_rating'])
df_without_column_for_convert.head()
```

### Preprocessing
Mengubah data categorial ke numerik

```python
df_ratarata=pd.get_dummies(data['rata_rata_overdue'])
df_ratarata.head()
```

```python
df_kpr_aktif=pd.get_dummies(data['kpr_aktif'])
df_kpr_aktif
```

```python
df_new = pd.concat([df_without_column_for_convert, df_kpr_aktif,df_ratarata], axis=1)
df_new.head()
```

```python
df_risk_rating = pd.DataFrame(data, columns = ['risk_rating'])
df_risk_rating.head()
```

```python
df_without_risk_rating=df_new.drop(['risk_rating'], axis=1)
```

```python
df_new2 = pd.concat([df_without_risk_rating,df_risk_rating], axis=1)
df_new2
```

### Normalisasi

```python
from sklearn.preprocessing import MinMaxScaler
df_for_minmax_scaler=pd.DataFrame(data, columns = ['pendapatan_setahun_juta','durasi_pinjaman_bulan','jumlah_tanggungan'])
df_for_minmax_scaler.to_numpy()
scaler = MinMaxScaler()
df_hasil_minmax_scaler=scaler.fit_transform(df_for_minmax_scaler)
```

```python
df_hasil_minmax_scaler = pd.DataFrame(df_hasil_minmax_scaler,columns = ['pendapatan_setahun_juta','durasi_pinjaman_bulan','jumlah_tanggungan'])
df_hasil_minmax_scaler
```

```python
df_drop_column_for_minmaxscaler=df_new2.drop(['pendapatan_setahun_juta','durasi_pinjaman_bulan','jumlah_tanggungan'], axis=1)
df_drop_column_for_minmaxscaler
```

```python
df_new3 = pd.concat([df_drop_column_for_minmaxscaler,df_hasil_minmax_scaler], axis=1)
df_new3
```

```python
df_new3_without_risk_rating=df_new3.drop(['risk_rating'], axis=1)
df_new3_without_risk_rating
```

### Data yang akan digunakan

```python
df_new4 = pd.concat([df_new3_without_risk_rating,df_risk_rating], axis=1)
df_new4
```

## Implementasi Metode Naive Bayes

```python
from sklearn.metrics import make_scorer, accuracy_score,precision_score
from sklearn.metrics import accuracy_score ,precision_score,recall_score,f1_score
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import KFold,train_test_split,cross_val_score
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
```

```python
X=df_new4.iloc[:,1:10].values
y=df_new4.iloc[:,11].values
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
gaussian = GaussianNB()
gaussian.fit(X_train, y_train)
Y_pred = gaussian.predict(X_test) 
accuracy_nb=round(accuracy_score(y_test,Y_pred)* 100, 2)
acc_gaussian = round(gaussian.score(X_train, y_train) * 100, 2)

cm = confusion_matrix(y_test, Y_pred)
accuracy = accuracy_score(y_test,Y_pred)
precision =precision_score(y_test, Y_pred,average='micro')
recall =  recall_score(y_test, Y_pred,average='micro')
f1 = f1_score(y_test,Y_pred,average='micro')
print('Confusion matrix for Naive Bayes\n',cm)
print('accuracy_Naive Bayes: %.3f' %accuracy)
print('precision_Naive Bayes: %.3f' %precision)
print('recall_Naive Bayes: %.3f' %recall)
print('f1-score_Naive Bayes : %.3f' %f1)
```

## Implementasi Metode K-NN

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 25)
knn.fit(X_train,y_train)
Y_pred = knn.predict(X_test)[0:]
accuracy_nb=round(accuracy_score(y_test,Y_pred)* 100, 2)
acc_neigh = round(knn.score(X_train, y_train) * 100, 2)

cm = confusion_matrix(y_test, Y_pred)
accuracy = accuracy_score(y_test,Y_pred)
precision =precision_score(y_test, Y_pred,average='micro')
recall =  recall_score(y_test, Y_pred,average='micro')
f1 = f1_score(y_test,Y_pred,average='micro')
print('Confusion matrix for K-NN\n',cm)
print('accuracy_K-NN: %.3f' %accuracy)
print('precision_K-NN: %.3f' %precision)
print('recall_K-NN: %.3f' %recall)
print('f1-score_K-NN : %.3f' %f1)
```

## Implementasi Metode Decision Tree

```python
from sklearn import tree

iristree = tree.DecisionTreeClassifier(max_depth=3, criterion='gini', random_state=0)
iristree.fit(X_train, y_train)
Y_pred = iristree.predict(X_test) 
accuracy_nb=round(accuracy_score(y_test,Y_pred)* 100, 2)
acc_gaussian = round(gaussian.score(X_train, y_train) * 100, 2)

cm = confusion_matrix(y_test, Y_pred)
accuracy = accuracy_score(y_test,Y_pred)
precision =precision_score(y_test, Y_pred,average='micro')
recall =  recall_score(y_test, Y_pred,average='micro')
f1 = f1_score(y_test,Y_pred,average='micro')
print('Confusion matrix for Decision Tree\n',cm)
print('accuracy_Decision Tree: %.3f' %accuracy)
print('precision_Decision Tree: %.3f' %precision)
print('recall_Decision Tree: %.3f' %recall)
print('f1-score_Decision Tree : %.3f' %f1)
```
