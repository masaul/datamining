```python
from matplotlib import rcParams, cycler
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math
plt.ion()
```

```python
dataset = pd.read_csv('https://gist.githubusercontent.com/netj/8836201/raw/6f9306ad21398ea43cba4f7d537619d0e07d5ae3/iris.csv')
dataset
```

```python
dataset.head()
```

```python
def jarakData(i, j):
  P = 3
  M = 0
  for data in ['race/ethnicity', 'parental level of education', 'lunch' ]:
    if dataset[data][i] == dataset[data][j]:
      M = M + 1
  return (P-M)/P
```

```python
print(jarakData(0,1))
print(jarakData(0,2))
print(jarakData(0,3))
```

## Atribut Numerik

```python
def jarakData2(i, j, h):
  temp = 0
  for dataColomn in ['math score', 'reading score', 'writing score']:
    result = math.pow(abs((dataset[dataColomn][i]) - (dataset[dataColomn][j])), h)
  return result**(1/h)
```

```python
print(jarakData2(0,1,1))
print(jarakData2(0,2,1))
print(jarakData2(0,3,1))
```

```python
print(jarakData2(0,1,3))
print(jarakData2(0,2,3))
print(jarakData2(0,3,3))
```
