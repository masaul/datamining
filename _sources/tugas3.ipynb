# **Implementasi K-NN pada datase iris (Tugas 3)**

```python
import pandas as pd
import numpy as np
pd.options.mode.chained_assignment = None 
```

## Mendapatkan dataset iris

```python
dataset = pd.read_csv("https://gist.githubusercontent.com/netj/8836201/raw/6f9306ad21398ea43cba4f7d537619d0e07d5ae3/iris.csv")
dataset
```

## Membuat dataframe

```python
X = dataset.drop(columns=['variety'])
```

```python
y = dataset['variety'].values
y[0:5]
```

## Split dataset data uji dengan data latih

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1, stratify=y)
```

## Membangun model

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_train,y_train)
```

## Test model

```python
knn.predict(X_test)[0:5]
```

```python
knn.score(X_test, y_test)
```

```python
from sklearn.model_selection import cross_val_score
import numpy as np

knn_cv = KNeighborsClassifier(n_neighbors=3)

cv_scores = cross_val_score(knn_cv, X, y, cv=5)

print(cv_scores)
print('cv_scores mean:{}'.format(np.mean(cv_scores)))
```

```python
from sklearn.model_selection import GridSearchCV

knn2 = KNeighborsClassifier()

param_grid = {'n_neighbors': np.arange(1, 25)}

knn_gscv = GridSearchCV(knn2, param_grid, cv=5)

knn_gscv.fit(X, y)
```

## check nilai n_neighboars

```python
knn_gscv.best_params_
```

## check rata-rata nilai n_neighboard

```python
knn_gscv.best_score_
```
