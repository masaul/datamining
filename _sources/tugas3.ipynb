## Konsep k-NN
k-NN (k-Nearest-Neighbors) merupakan sebuah algoritma yang digunakan untuk menentukan sebuah label dari data labeh yang sudah tersedia.
k-NN ini memeliki 5 algoritma yang digunakan
1.   Menentukan nilai K
2.   Menghitung jarak antara data uji dengan data latih (menggunakan Minkowski Distance)
3. Mengurutkan jarak dari terkecil hingga terbesar (Ascending)
4. Mengambil data sebanyak K terdekat
5. Memilih kelas

Menghitung jarak dengan cara data latih - data uji pada setiap fitur/label yang digunakan. kemudian jumlahkan.

Jika jumlah hasil label tersebut sama, maka bisa juga dipilih salah satu.




```python
import pandas as pd
import numpy as np
pd.options.mode.chained_assignment = None 
```

## Mendapatkan dataset iris

```python
dataset = pd.read_csv("https://gist.githubusercontent.com/netj/8836201/raw/6f9306ad21398ea43cba4f7d537619d0e07d5ae3/iris.csv")
dataset
```

## Membuat dataframe
Label yang diguanakan adalah variety

```python
X = dataset.drop(columns=['variety'])
```

```python
y = dataset['variety'].values
y[0:29]
```

## Split dataset data uji dengan data latih
Ukuran data yang diguanakan adalah 20% dari data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1, stratify=y)
```

## Membangun model
K yang digunakan sebanyak 15

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 15)
knn.fit(X_train,y_train)
```

## Test model
Label yang diambil sebanyak 29 data yang sudah tersedia

```python
knn.predict(X_test)[0:29]
```

```python
knn.score(X_test, y_test)
```

```python
from sklearn.model_selection import cross_val_score
import numpy as np

knn_cv = KNeighborsClassifier(n_neighbors=3)

cv_scores = cross_val_score(knn_cv, X, y, cv=5)

print(cv_scores)
print('cv_scores mean:{}'.format(np.mean(cv_scores)))
```
