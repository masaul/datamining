# **Disimilarity (Tugas 1)**

## Konsep Disimilarity
Dalam peambangan data yang pengaplikasian yang sering digunakan adalah clustering. pada teknik ini perlu untuk menemukan seberapa mirip atau tidaknya data. Salah satu cara yaitu dengan menggunakan Dismilarity

Terdapat beberapa atribut yang dimiliki seperti 

1. Nominal / Categorial
2. Ordinal
3. Biner
4. Numerik 




Mengukur jarak (dissimilarity)
1.   Ambil data dari Kaggle atau GitHub
2.   Ukur jarak d(1,2), d(1,3), d(1,4) dari data tersebut

```python
import pandas as pd
import math
```

```python
dataset = pd.read_csv('https://raw.githubusercontent.com/jakobgross/Students-Performance-in-Exams-Kaggle-/master/StudentsPerformance.csv')
dataset.head()
```

## Atribut Categorial
Atribut yang menggunakan nama suatu hal atau kategori.

Rumus Atribut Cetogorial
$\frac {P-M}{P}$

```python
def jarakData(i,j):
  p = 4
  m = 0
  for col in ['race/ethnicity', 'parental level of education', 'test preparation course', 'lunch']:
    if dataset[col][i] == dataset[col][j]:
      m+=1
  return (p-m)/p
```

```python
print(jarakData(0, 1))
print(jarakData(0, 2))
print(jarakData(0, 3))
```

## Atribut Numerik
Atribut yang memiliki nilai integer atau nilai real.

```python
def jarakData2(i, j, h):
  temp = 0
  for dataColomn in ['math score', 'reading score', 'writing score']:
    result = math.pow(abs((dataset[dataColomn][i]) - (dataset[dataColomn][j])), h)
  return result**(1/h)
```

```python
print(jarakData2(0,1,1))
print(jarakData2(0,2,1))
print(jarakData2(0,3,1))
```

```python
print(jarakData2(0,1,3))
print(jarakData2(0,2,3))
print(jarakData2(0,3,3))
```
